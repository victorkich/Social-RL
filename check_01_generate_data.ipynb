```python
import numpy as np
import matplotlib.pyplot as plt
import os
from IPython import display
```

```python
DIR_NAME = './data/rollout/'
file = os.listdir(DIR_NAME)[2]
print(file)
rollout = np.load(DIR_NAME + file, allow_pickle=True)
obs_data = rollout['obs']
action_data = rollout['action']
reward_data = rollout['reward']

print(obs_data.shape)
```

```python
import numpy as np

fig, axes = plt.subplots(nrows=5, ncols=4, figsize=(64,64), sharex=True, sharey=True)
plt.subplots_adjust(hspace = 0.4)
idx = 99
for x in range(5):
    for y in range(4):
        obs = obs_data[idx]

        # Assegura que 'obs' é um array do NumPy e tem o tipo de dado correto
        if not isinstance(obs, np.ndarray):
            obs = np.array(obs)
        if obs.dtype != np.float32 and obs.dtype != np.uint8:
            obs = obs.astype(np.float32)

        # Verifica o formato de 'obs'
        if obs.ndim != 3 or obs.shape[-1] != 3:
            # Lidar com o erro de formato aqui
            pass

        if obs.dtype == np.float32:
            # Normaliza para o intervalo [0, 1] se necessário
            if obs.max() > 1.0:
                obs /= 255.0

        axes[x, y].imshow(obs)
```
